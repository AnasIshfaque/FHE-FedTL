```python
import os
import h5py
import numpy as np

import socket
import struct
import pickle

import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision
from torchvision import datasets,transforms,models
import torch.optim as optim

from torch.utils.data import Dataset, DataLoader

import time
import copy

from tqdm import tqdm
```

```python
root_path = '../../datasets/THB_splitted'
```

```python
client_orders = [0, 1, 2]
users = 3
```

```python
num_traindata = 262 // users
print(num_traindata)
indices = list(range(262))
parts = []

for client_order in client_orders:
    lower_idx = num_traindata * client_order
    upper_idx = num_traindata * (client_order + 1)
    if (client_order+1 == users):
        upper_idx += 1
    part_size = upper_idx - lower_idx
    print(part_size)
    parts.append(indices[lower_idx : upper_idx])

for part in parts:
    print(part)
```

```python
mean = np.array([0.485,0.456,0.406])
std = np.array([0.229,0.224,0.225])
transform = transforms.Compose([transforms.RandomResizedCrop(224),
                                transforms.RandomHorizontalFlip(),
                                transforms.ToTensor(),
                                transforms.Normalize(mean,std)])

trainset = datasets.ImageFolder(os.path.join(root_path,'train'), transform)
```

```python
from torch.utils.data import Subset
trainset_subs = []

for part_tr in parts:
    trainset_subs.append(Subset(trainset, part_tr))

for trainset_sub in trainset_subs:
    print(len(trainset_sub))
```

```python
dataloaders = []

for trainset_sub in trainset_subs:
    dataloaders.append(torch.utils.data.DataLoader(trainset_sub, batch_size=4, shuffle=True, num_workers=0))

for loader in dataloaders:
    print(len(dataloaders))
```
