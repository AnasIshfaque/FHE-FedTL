```python
import torch
import torchvision
from torchvision import datasets,transforms,models
from torchvision.models import efficientnet_v2_s, EfficientNet_V2_S_Weights
```

```python
model = models.efficientnet_v2_s(weights=EfficientNet_V2_S_Weights.DEFAULT)
model
```

```python
# Access classifier module (might vary slightly based on implementation)
classifier = model._modules['classifier']

# Get number of input features from existing layer
in_features = classifier[1].in_features

# Replace final layer with new layer for 2 outputs
classifier[1] = torch.nn.Linear(in_features, 2)

model
```
